```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results_file = "training_logs/training_metrics_a2c_20251217_124749.json"

# load the json

results = json.load(open(results_file))
```

```python
results.keys()
```

```python
# plot the full results and moving average of 'episode_rewards', 'episode_lengths', 'episode_losses', 'episode_max_lengths'

plot_values = ['episode_rewards', 'episode_lengths', 'episode_losses', 'episode_max_lengths', 'episode_reward_per_length']


results['episode_reward_per_length'] = [x / y for x, y in zip(results['episode_rewards'], results['episode_lengths'])]

for key in plot_values:
    plt.figure(figsize=(10, 5))
    metric_df = pd.DataFrame(results[key])
    metric_df['rolling_mean'] = metric_df[0].rolling(5).mean()
    metric_df.plot()
    # turn off legend
    plt.legend([])
    plt.title(key)
    plt.show()
# plot the episode_reward_per_length

```
